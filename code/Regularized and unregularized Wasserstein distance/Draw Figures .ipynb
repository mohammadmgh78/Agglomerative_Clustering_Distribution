```python
#read the data that we have generated and see the plots, other wise run the code above. 
# Open and read the txt file
with open("../../data/regularized.txt", "r") as file:
    data = file.read()

# Convert the string from the file into a list of lists
times_list_reg_total = eval(data)

with open("../../data/unregularized.txt", "r") as file:
    data = file.read()
# Convert the string from the file into a list of lists
times_list_unreg_total = eval(data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x_axis = [10,100,200,300,600,900,1200,1600,2000,2400,3000]
data1 = times_list_reg_total  # First dataset (Regularized)
data2 = times_list_unreg_total  # Second dataset (Unregularized)

# Calculate means for both datasets
means1 = np.mean(data1, axis=1)
means2 = np.mean(data2, axis=1)

# Plotting
plt.figure(figsize=(12, 8))  # Increase figure size

# Plot the first dataset (Regularized)
plt.plot(x_axis, means1, marker='o', linestyle='-', color='b', markersize=8, linewidth=2, label='Regularized')  # Regularized line

# Plot the second dataset (Unregularized)
plt.plot(x_axis, means2, marker='s', linestyle='--', color='r', markersize=8, linewidth=2, label='Unregularized')  # Unregularized line

# Set xticks and labels based on x_axis, rotate the labels by 45 degrees
plt.xticks(x_axis, [f'{i}' for i in x_axis], rotation=45, fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Support Size', fontsize=22)
plt.ylabel('Time (seconds)', fontsize=22)

# Add legend
plt.legend(fontsize=22, frameon=False)

# Remove the upper and right borders
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Auto-adjust layout to prevent overlap
plt.tight_layout()
plt.savefig('../../results/regularized vs unregularized.png', dpi=300, bbox_inches='tight', transparent=True)
# Show the plot
plt.show()
```
