```python
from sklearn.cluster import KMeans
import numpy as np

import numpy as np

# Initialize an empty list to store the points
possible_points = []


for x1 in range(47):  # 0 to 47
    for x2 in range(47-x1):  # 0 to 47
        for x3 in range(47-x1-x2):  # 0 to 47
            if x1 + x2 + x3 == 46:
                possible_points.append([x1, x2, x3])
                
                



# Convert the list of points to a NumPy matrix
points_matrix = np.array(possible_points)

points = points_matrix

# Number of clusters
k = 100

# Initialize KMeans
kmeans = KMeans(n_clusters=k)

# Fit the model
kmeans.fit(points)

# Get cluster labels
labels = kmeans.labels_

# Get cluster centroids
centroids = kmeans.cluster_centers_
integer_matrix = centroids.astype(int)
```

```python
def adjust_matrix(matrix):
    new_matrix = np.copy(matrix)
    for i in range(matrix.shape[0]):
        row_sum = np.sum(new_matrix[i])
        while row_sum != 46:
            if row_sum <46:
                col_index = np.random.randint(3)
                new_matrix[i, col_index] += 1
                row_sum = np.sum(new_matrix[i])
            else:
                col_index = np.random.randint(3)
                new_matrix[i, col_index] -= 1
                row_sum = np.sum(new_matrix[i])                

    return new_matrix
integer_matrix = adjust_matrix(integer_matrix)
```
