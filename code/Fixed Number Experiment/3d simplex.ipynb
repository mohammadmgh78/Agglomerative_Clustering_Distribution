```python
import numpy as NP
import matplotlib.pyplot as P
import matplotlib.ticker as MT
import matplotlib.lines as L

def plotSimplex(points_list, labels_list, fig=None,
                vertexlabels=['(49,0,0)', '(0,49,0)', '(0,0,49)'],
                colors=['blue', 'red', 'green', 'orange', 'purple'],  # Add five distinct colors
                markers=['o', 's', 'D', '^', 'P'],  # Different marker shapes for each cluster
                **kwargs):
    """
    Plot multiple sets of Nx3 points arrays on the 3-simplex
    (with optionally labeled vertices)

    kwargs will be passed along directly to matplotlib.pyplot.scatter
    Returns Figure, caller must .show()
    """
    if fig is None:
        fig = P.figure(figsize=(8, 8))  # Increased figure size for better spacing

    ax = fig.add_subplot(111, aspect='equal')

    # Draw the triangle
    l1 = L.Line2D([0, 0.5, 1.0, 0],  # xcoords
                  [0, NP.sqrt(3) / 2, 0, 0],  # ycoords
                  color='k')
    ax.add_line(l1)
    ax.xaxis.set_major_locator(MT.NullLocator())
    ax.yaxis.set_major_locator(MT.NullLocator())

    # Draw vertex labels with additional padding to avoid overlap
    ax.text(-0.1, -0.05, vertexlabels[0], ha='center')
    ax.text(1.1, -0.05, vertexlabels[1], ha='center')
    ax.text(0.5, NP.sqrt(3) / 2 + 0.08, vertexlabels[2], ha='center')

    # Project and draw the actual points with legends
    for i, points in enumerate(points_list):
        projected = projectSimplex(points)
        ax.scatter(projected[:, 0], projected[:, 1], color=colors[i], marker=markers[i], label=labels_list[i], **kwargs)

    # Add legend with manual ordering
    handles, labels = ax.get_legend_handles_labels()
    # Move "Other clusters" to the end of the list
    if 'Other clusters' in labels:
        other_index = labels.index('Other clusters')
        # Move "Other clusters" to the end
        handles.append(handles.pop(other_index))
        labels.append(labels.pop(other_index))

    ax.legend(handles, labels, loc='upper right', bbox_to_anchor=(1.2, 1), borderaxespad=0.)  # Move legend outside plot

    # Remove the axis
    ax.axis('off')

    # Leave some buffer around the triangle for vertex labels and better layout
    ax.set_xlim(-0.2, 1.2)
    ax.set_ylim(-0.2, 1.2)

    # Adjust layout to avoid overlaps
    P.tight_layout(pad=3)

    return fig

def projectSimplex(points):
    """
    Project probabilities on the 3-simplex to a 2D triangle

    N points are given as N x 3 array
    """
    # Normalize points to the sum of 49
    points = points / 49

    # Convert points one at a time
    tripts = NP.zeros((points.shape[0], 2))
    for idx in range(points.shape[0]):
        # Init to triangle centroid
        x = 1.0 / 2
        y = 1.0 / (2 * NP.sqrt(3))
        # Vector 1 - bisect out of lower left vertex
        p1 = points[idx, 0]
        x = x - (1.0 / NP.sqrt(3)) * p1 * NP.cos(NP.pi / 6)
        y = y - (1.0 / NP.sqrt(3)) * p1 * NP.sin(NP.pi / 6)
        # Vector 2 - bisect out of lower right vertex
        p2 = points[idx, 1]
        x = x + (1.0 / NP.sqrt(3)) * p2 * NP.cos(NP.pi / 6)
        y = y - (1.0 / NP.sqrt(3)) * p2 * NP.sin(NP.pi / 6)
        # Vector 3 - bisect out of top vertex
        p3 = points[idx, 2]
        y = y + (1.0 / NP.sqrt(3) * p3)

        tripts[idx, :] = (x, y)

    return tripts


if __name__ == '__main__':

    # Define example points (replace with actual datasets)
    testpoints_1 = integer_matrix
    testpoints_2 = cluster_4
    testpoints_3 = first_cheapest
    testpoints_4 = second_cheapest
    testpoints_5 = sixth_cheapest

    # Do scatter plot with five sets of points using different shapes for each cluster
    fig = plotSimplex([testpoints_1, testpoints_2, testpoints_3, testpoints_4, testpoints_5], 
                      labels_list=['Other clusters', 'Cluster 4', 'Cheapest in Cluster 4',
'Second Cheapest in Cluster 4',
'Sixth Cheapest in Cluster 4'], 
                      colors=['pink', 'orange', '#52a447', 'red', 'blue'], 
                      markers=['o', 'D', '^', '*', 's'], s=40)  # Specify markers for different clusters
    
    P.savefig(f'../../results/plot_input_3dsimplex_with_legend.png', format='png', dpi=500)
    P.show()
```
