```python
from sklearn.metrics import adjusted_rand_score
import numpy as np
# Function to read cluster assignments from a text file
def read_cluster_assignments(file_path):
    with open(file_path, 'r') as f:
        clusters = [int(line.strip()) for line in f]
    return clusters
```

```python
ari_list_crn = []
# Paths to the text files containing cluster assignments
for i in range(100):
    file_path_1 = f'../../data/clusters_output_{i}.txt'
    file_path_2 = f'../../data/clusters_output_true_0.txt'



    # Read cluster assignments from each file
    clusters_1 = read_cluster_assignments(file_path_1)
    clusters_2 = read_cluster_assignments(file_path_2)
    # Calculate the Adjusted Rand Index (ARI)
    ari = adjusted_rand_score(clusters_1, clusters_2)
    ari_list_crn.append(ari)
```

```python
ari_list_independent = []
# Paths to the text files containing cluster assignments
for i in range(100):
    file_path_1 = f'../../data/clusters_output_independent_{i}.txt'
    file_path_2 = f'../../data/clusters_output_true_0.txt'



    # Read cluster assignments from each file
    clusters_1 = read_cluster_assignments(file_path_1)
    clusters_2 = read_cluster_assignments(file_path_2)
    # Calculate the Adjusted Rand Index (ARI)
    ari = adjusted_rand_score(clusters_1, clusters_2)
    ari_list_independent.append(ari)
   
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example Data (Replace these lists with your actual data)

# Plot KDE for each dataset with clipping and bandwidth adjustment
sns.kdeplot(ari_list_independent, fill=True, alpha=0.5, label='Independent', clip=(0, 1), bw_adjust=0.4)
sns.kdeplot(ari_list_crn, fill=True, alpha=0.5, label='CRN', clip=(0, 1), bw_adjust=0.4)

# Add titles and labels with larger font sizes
plt.xlabel('ARI', fontsize=16)
plt.ylabel('Density', fontsize=16)

# Customize the size of the ticks on both axes
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add legend in the upper left corner
plt.legend(loc='upper left', fontsize=12)

# Save the figure
plt.savefig('../../results/ARI_fig.png', format='png', dpi=300)

# Display the plot
plt.show()
```
